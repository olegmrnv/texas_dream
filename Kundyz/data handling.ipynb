```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
school_ratings_df = pd.read_csv("data/modified/combine_schools_zip_geo.csv")
school_ratings_df.head()
```

```python
school_ratings_df = school_ratings_df[['SCHOOL_TYPE', 'Campus Name',
                                       'District Name', 'Total Score',
                                       'Overall Grade', 'County Name']]
school_ratings_df.head()
```

```python
school_ratings_df = school_ratings_df.rename(columns={'SCHOOL_TYPE': 'School Type'})
school_ratings_df.head()
```

```python
school_ratings_df['County Name'] = school_ratings_df['County Name'].str.upper()
school_ratings_df.head()
```

```python
engine = create_engine('sqlite:///tx_school_data.sqlite', echo=False)
```

```python
school_ratings_df.to_sql('School_Ratings', con=engine, if_exists='replace')
```

```python
engine.execute('SELECT * FROM School_Ratings').fetchall()
```
