```python
import pandas as pd
from sqlalchemy import create_engine

df_unemp = pd.read_csv("Data/modified/Unemployment_Rate.csv")
df_income = pd.read_csv("Data/modified/Per_Capita_Personal_Income.csv")
```

```python
df_income.head()
```

```python
df_income = df_income[["Region Name", "2017"]]
```

```python
df_income.head()
```

```python
df_income['StateInitial'] = df_income['Region Name'].str[-2:]
df_income.head()
```

```python
df_income.head()
```

```python
df_income = df_income.loc[df_income["StateInitial"] == "TX", :]
df_income.head()
```

```python
df_income.reset_index(inplace=True)
df_income.head()
```

```python
df_income["Region_Name"] = df_income["Region Name"].str[:-11]
df_income["Region_Name"] = df_income["Region_Name"].str.upper()
df_income["Income"] = df_income["2017"]
```

```python
df_income=df_income[["Region_Name","Income"]]
```

```python
df_income
```

```python
engine = create_engine('sqlite:///tx_data.sqlite', echo=False)
```

```python
df_income.to_sql('Per_Capita_Personal_Income', con=engine)
df_unemp.to_sql('Unemployment_Rate', con=engine)
```

```python
engine.execute("SELECT * FROM Per_Capita_Personal_Income").fetchall()
```
