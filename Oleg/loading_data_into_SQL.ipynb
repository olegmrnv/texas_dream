```python
import pandas as pd
from sqlalchemy import create_engine

df_unemp = pd.read_csv("Data/modified/Unemployment_Rate.csv")
df_income = pd.read_csv("Data/modified/Per_Capita_Personal_Income.csv")
```

```python
df_income.head()
```

```python
df_income = df_income[["Region Name", "2017"]]
```

```python
df_income.head()
```

```python
df_income['StateInitial'] = df_income['Region Name'].str[-2:]
df_income.head()
```

```python
df_income.head()
```

```python
df_income = df_income.loc[df_income["StateInitial"] == "TX", :]
df_income.head()
```

```python
df_income.reset_index(inplace=True)
df_income.head()
```

```python
df_income["County"] = df_income["Region Name"].str[:-11]
df_income["County"] = df_income["County"].str.upper()
df_income["Income"] = df_income["2017"]
```

```python
df_income=df_income[["County","Income"]]
```

```python
df_income.head()
```

```python
df_unemp.head()
```

```python
del df_unemp['Series ID']
```

```python
del df_unemp['Region Code']
df_unemp.head()
```

```python
df_unemp['mean'] = df_unemp.mean(axis=1)
df_unemp.head()
```

```python
df_unemp['StateInitial'] = df_unemp['Region Name'].str[-2:]
df_unemp = df_unemp.loc[df_unemp["StateInitial"] == "TX", :]
df_unemp.reset_index(inplace=True)
df_unemp["County"] = df_unemp["Region Name"].str[:-11]
df_unemp["County"] = df_unemp["County"].str.upper()
df_unemp["Unemployment"] = df_unemp["mean"]
```

```python
df_unemp=df_unemp[['County','Unemployment']]
df_unemp.head()
```

```python
engine = create_engine('sqlite:///tx_data.sqlite', echo=False)
```

```python
df_income.to_sql('Per_Capita_Personal_Income', con=engine, if_exists='replace')
df_unemp.to_sql('Unemployment_Rate', con=engine, if_exists='replace')
```

```python
engine.execute("SELECT * FROM Per_Capita_Personal_Income").fetchall()
```
